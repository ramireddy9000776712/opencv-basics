```
import cv2
import numpy as np
from matplotlib import pyplot as plt

# load the image
img=cv2.imread('panda.jpg',0)

plt.imshow(img),plt.title('original image')
plt.show()

# apply global thresholding
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(th1),plt.title('simple threshold image')
plt.show()
# apply mean thresholding
# the function calculates the mean of a 11x11 neighborhood area for each pixel
# and subtract 2 from the mean
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 11, 2)
plt.imshow(th2),plt.title('mean threshold image')
plt.show()
# apply Gaussian thresholding
# the function calculates a weights sum by using a 11x11 Gaussian window
# and subtract 2 from the weighted sum.
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(th3),plt.title('gaussian threshold image')
plt.show()
# display the processed images
#titles = ['Original Image', 'Global Thresholding (v = 127)',
         # 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
#images = [img, th1, th2, th3]

#for i in range(4):
   # plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
   # plt.title(titles[i])
   # plt.xticks([]), plt.yticks([])
#plt.show()
```

```
=>IMAGE BLURRING OR SMOOTHING OR FILTERING:
1.TO REMOVE THE NOISE ON AN IMAGE


=>IMAGE NOISE:
1.capturing pictures from low quality camera ar night results in a grainy picture that is noise.


=>WAYS TO FILTER AN IMAGE:
1.images can be filtered by low pass filter or high pass filter.
2.low pass filter (LPF) can help us to remove the noise from the image or blur the image.
3.High pass filter(HPF)can help us to find the edges in an image

=>HOW IMAGE FILTERING WORKS:
1.we need something called as KERNEL.kernel is nothing but the matrix
2.This kernel is convolved with the image to get the resultant filtered image


=>HOW CONVOLUTION WORKS:




```

```
import numpy as np

image=cv2.imread('panda.jpg')
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
matrix=np.ones((5,5),np.float32)/25     #float32 is a data type and ones(5,5)is a 5x5 matrix contains all ones
new_image=cv2.filter2D(img,-1,matrix)   # -1 tells the depth ie depth of new image is same as original image
plt.imshow(image),plt.title('image in BGR')
plt.show()
plt.imshow(img),plt.title('image in RGB')
plt.show()
plt.imshow(new_image),plt.title('general filtering image')
plt.show()

# AVERAGING (LOW PASS FILTERING)

blur = cv2.blur(img,(10, 10))         # (10,10) IS A MATRIX OF SIZE and 
                                        #this 10x10 matrix is placed on a pixel and take the average of all pixels under this matrix ,finally that pixel is replaced with the average value
plt.imshow(blur), plt.title('Blurred image by averaging')
plt.show()

#GAUSSIAN FILTERING

#1.A Gaussian filter is a linear filter. It's usually used to blur the image or to reduce noise.The Gaussian filter alone will blur edges and reduce contrast. 
#   The Median filter is a non-linear filter that is most commonly used as a simple way to reduce noise in an image.
gblur = cv2.GaussianBlur(img,(5,5),0)
plt.imshow(gblur),plt.title('blurred image by gaussian method')
plt.show()
```
